```python
import re  # Importing the re module

def trim_model(x):
    match = re.search(r"^(.*?)\s*\(", x)  # Regex to find text before optional spaces and parenthesis
    if match:
        return match.group(1).strip()  # Return the captured group and strip extra whitespace
    return x  # Return original value if no match is found

df["model_trimmed"] = df["model_trimmed"].apply(trim_model)
```
